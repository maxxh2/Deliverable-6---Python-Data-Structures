```python
# Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Loading iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Dropping the species column
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
# Adding columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']

iris.head()
```

```python
# Creating new dataframe

mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

summary_dict = {'mean': mean, 'std': std, 'min': min, 'max': max} 

summary_df = pd.DataFrame(data = summary_dict)

summary_df = pd.DataFrame(summary_dict, index = ['sepal_length',
                                                 'sepal_width',
                                                 'petal_length',
                                                 'petal_width',
                                                 'sepal_sum',
                                                 'petal_sum'])


summary_df
```
